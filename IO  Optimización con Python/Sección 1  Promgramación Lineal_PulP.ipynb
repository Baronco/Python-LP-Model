```python
import pulp as p
import pandas as pd
```

```python
modelo = p.LpProblem('Produccion_X_Y',p.LpMaximize)
```

```python
x = p.LpVariable('Produccion_Semanal_X',lowBound=0, cat=p.LpInteger)
y = p.LpVariable('Produccion_Semanal_Y',lowBound=0, cat=p.LpInteger)
```

```python
# FO
modelo += 40*x + 30*y
```

```python
# Constrains
modelo += x <= 40
# capacidad horas de tarea A
modelo += x + y <= 80
# capacidad horas de tarea B
modelo += 2*x + y <= 100
```

```python
modelo
```

```python
# solution
modelo.solve()
```

```python
print(f'Número de unidades semanales a producir de X: ', x.varValue)
print(f'Número de unidades semanales a producir de Y: ', y.varValue)
print('Contribución Máxima: ', p.value(modelo.objective))
```

Precio sombra no se puede obtener cuando las variables tienen como restricción ser Constantes

```python
# analisis de sensibilidad
sensibilidad = [
    {
        'Restricción':i,
        'Precio Sombra':j.pi,
        'Holgura':j.slack
    }
    for i,j in modelo.constraints.items()
]

print(pd.DataFrame(sensibilidad))
```
